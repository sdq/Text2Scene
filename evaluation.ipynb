### split images in dir to train and test

```python
from tools.common import getBase
import glob
import random
import os
import shutil
URIs = [getBase(path) for path in glob.glob('text/*.txt')]
random.shuffle(URIs)
print(len(URIs))
train = URIs[:70]
val = URIs[70:]

def train_val_split(src, train, val, ext):
    train_dir = '%s_train' % src
    val_dir = '%s_val' % src
    if os.path.isdir(train_dir):
        shutil.rmtree(train_dir)
    os.mkdir(train_dir)
    if os.path.isdir(val_dir):
        shutil.rmtree(val_dir)
    os.mkdir(val_dir)
    for name in train:
        copyfile('%s/%s%s' % (src, name, ext),
                 '%s/%s%s' % (train_dir, name, ext))
    for name in val:
        copyfile('%s/%s%s' % (src, name, ext),
                 '%s/%s%s' % (val_dir, name, ext))
train_val_split('images', train, val, '.svg')
train_val_split('text', train, val, '.txt')
```

### metric on the entire set: train

```python
from models.pipeline import Translate
translate = Translate(img_dir="images_train")
```

```python
from tools.containers import Picture, Description
from tools.common import getBase
import glob
# from models.metric import WeightedMeanIOU
import random

def random_gen(translate):
    return random.choice(list(translate.ground.layerbase.pic_vocab_))

def avg_list_of_dict(ld):
    keys = ld[0].keys()
    avg_ = {}
    for key in keys:
        avg_[key] = sum([d[key] for d in ld]) / len(ld)
    return avg_

# metric = WeightedMeanIOU()
simis = []
for name in [getBase(path) for path in glob.glob('text_train/*.txt')]:
    print(' -------------------------- ')
    print(name)
    # print(Description('text_train/%s.txt' % name))
    true_pic = Picture('images_train/%s.svg' % name)
    # gen_pic = translate(path='text_train/%s.txt' % name, verbose=False)
    gen_pic = random_gen(translate)
    # print(true_pic)
    # print(gen_pic)
    simi = translate.metric.picture_simi(true_pic, gen_pic,
                                         with_bg=False)
    # print(simi)
    simis.append(simi)
# print(sum(simis)/len(simis))
train_simi_dict = avg_list_of_dict(simis)
```
