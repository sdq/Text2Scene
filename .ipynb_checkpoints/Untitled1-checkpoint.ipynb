```python
from models.dataset import Dataset, DatasetReality
from models.pipeline import prep_data, train_classifier
from ana.visual import STAT, ROC, FEAT

import numpy as np
import random
np.random.seed(7)
random.seed(7)

import warnings
from sklearn.linear_model import LogisticRegression
warnings.simplefilter('ignore')

## for evaluation
import matplotlib.pyplot as plt
```

```python
# dataset, X, y, pairs, index = prep_data(Dataset, shuffle_interval=3)
dataset = prep_data(Dataset, test=0.1, shuffle=True, seed=None,
                    text_idf=True, categ_idf=False,
                    cross_simi=True,
                    norm_simi=True, suppress_freq=1)
```

```python
# plt.figure(figsize=(10,6))
fig, ax = plt.subplots(1,3,figsize=(10,6))

titles = ['Highest 20 pictorial features',
          '20 ',
          'Highest 20 joint features']
for p, encoder in enumerate([dataset.img_encoder,
                             dataset.txt_encoder,
                             dataset.joint_encoder]):
#     # plt.subplot(1,3,1)
#     feats, vals = zip(*list(filter(lambda x: x[1] != 0, sorted(dataset.img_encoder.getFeatureValues().items(), key=lambda x: x[1])[::-1]))[:20])
#     ax[0].barh(feats, vals)
#     ax[0].invert_yaxis()

# # plt.subplot(1,3,2)
# feats, vals = zip(*list(filter(lambda x: x[1] != 0, sorted(dataset.joint_encoder.getFeatureValues().items(), key=lambda x: x[1])[::-1]))[:20])
# ax[1].barh(feats, vals)
# ax[1].invert_yaxis()
# ax[1].set_yticks('')
# for i, txt in enumerate(feats):
#     # print(txt)
#     ax[1].text(0.01, bar3[i].get_y()+0.8-0.2, txt, color='k')

    # plt.subplot(1,3,3)
    if p == 1:
        feats, vals = zip(*list(sorted(encoder.getFeatureValues().items(), key=lambda x: x[1]))[:20])
    else:
        feats, vals = zip(*list(sorted(encoder.getFeatureValues().items(), key=lambda x: x[1]))[-20:])
    bar = ax[p].barh(feats, vals, color='orange')
    if p == 1:
        ax[p].invert_yaxis()
    ax[p].set_yticks('')
    for i, txt in enumerate(feats):
        # print(txt)
        # pass
        ax[p].text(0.01, bar[i].get_y()+0.8-0.4, txt,
                   color='k', fontsize=9)
    # ax[2].set_facecolor("grey")
    ax[p].set_title(titles[p])
plt.savefig('results/feat_val')
plt.show()
```
